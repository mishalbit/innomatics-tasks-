# Calculation of Mean, Median and Mode (Measures of Central tendancy)

#Monthly HH Income

```python
Mthly_HH_Income = Dataset['Mthly_HH_Income'].values
print ('Maximum Monthly HH Income would be : ', Mthly_HH_Income.max())
print ('Minimum Monthly HH Income would be : ', Mthly_HH_Income.min())
print ('Average Monthly HH Income would be : ', Mthly_HH_Income.mean())
print ('Standard Deviation of Monthly HH Income would be : ', Mthly_HH_Income.std())
print ('Variance of Monthly HH Income would be : ', Mthly_HH_Income.var())
print('')
print ('Median of Monthly HH Income would be : ', np.median(Mthly_HH_Income))
print ('Quartile 50 % value would be : ', scoreatpercentile(Mthly_HH_Income, 50))
print ('Quartile 75% value would be : ', scoreatpercentile(Mthly_HH_Income, 75))
print ('Mode of Monthly HH Income would be : ', stats.mode(Mthly_HH_Income))
```

```python
def QQ_plot(data, measure):
    fig = plt.figure(figsize = (20,7))
    (mu, sigma) = norm.fit(data)
    
    fig1 = fig.add_subplot(121)
    sns.distplot(data, fit = norm)
    fig1.set_title(measure + ' Distribution ( mean = {:.2f} and standard deviation = {:.2f} )'.format(mu, sigma), loc = 'center')
    fig1.set_xlabel(measure)
    fig1.set_ylabel('Frequency')

    plt.tight_layout()
    plt.show()
```

```python
QQ_plot(Dataset.Mthly_HH_Income, 'Mthly_HH_Income')
```

# Understanding our data through QQ-Plots
The quantile-quantile (q-q) plot is a graphical technique for determining if two data sets come from populations with a common distribution. A q-q plot is a plot of the quantiles of the first data set against the quantiles of the second data set.
